Define the getNearbyVenues function.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

That's a lot of different types of venues. Let's create a more concentrated list. Looking at this data from a tourist's perspective, they might be interested in coffee shops, restaurants, places of cluture, such ar museeums or theaters, parks and lakes and places to shop.

Let's create a column for each of these categories.

```python
#set conditions for Coffe shop or cafe
conditionsC = [
    (venues['Venue Category'] == "Café") |(venues['Venue Category'] == "Coffee Shop")
    ]
#set conditions for museums, theaters and art
conditionsM = [
    (venues['Venue Category'].str.contains("Museum", na=False)) | (venues['Venue Category'].str.contains("Theater", na=False)) | (venues['Venue Category'].str.contains("Art", na=False))
    ]
#set conditions for parks and lakes
conditionsP = [
    (venues['Venue Category'].str.contains("Park", na=False)) |
    (venues['Venue Category'].str.contains("Garden", na=False)) |
    (venues['Venue Category'].str.contains("Lake", na=False))
    ]
#set conditions for restaurants and bars
conditionsR = [
    (venues['Venue Category'].str.contains("Restaurant", na=False))|
    (venues['Venue Category'].str.contains("Pizza", na=False))|
    (venues['Venue Category'].str.contains("Bar", na=False))
    ]
#set conditions for other shops and stores
conditionsS = [
    ((venues['Venue Category'].str.contains("Shop", na=False)) & (venues['Venue Category'] != "Coffee Shop")) | (venues['Venue Category'].str.contains("Store", na=False))
    ]
```

```python
venues['Coffee'] = np.transpose(np.where(conditionsC, 1, 0))
venues['Culture'] = np.transpose(np.where(conditionsM, 1, 0))
venues['Park'] = np.transpose(np.where(conditionsP, 1, 0))
venues['Restaurant'] = np.transpose(np.where(conditionsR, 1, 0))
venues['Store'] = np.transpose(np.where(conditionsS, 1, 0))
```
